```python
import pandas as pd
import os
```

```python
scraped_data = pd.read_pickle('../0 data/Scraped food recipes dataset')
```

```python
scraped_data.shape
```

```python
scraped_data.head()
```

```python
# using first 35 scraped recipes data
steps_data = scraped_data[["Recipe Name", 'Steps']][:35]
```

```python
steps_data
```

```python
for index in range(len(steps_data)):
    recipe_name = steps_data.iloc[index]["Recipe Name"]
    
    for step_index,step in enumerate(steps_data.iloc[index]["Steps"]):
        with open(f'../0 data/Dataturks Raw Files To Annotate/Recipe {recipe_name} {step_index}.txt', 'w') as input:
            input.write(step)
```
